Here we remove similar images/products using Elasticsearch.

```python
from elasticsearch import Elasticsearch
from image_match.elasticsearch_driver import SignatureES
from tqdm import tqdm_notebook as tqdm

es = Elasticsearch()
ses = SignatureES(es)
```

```python
import os
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

dir_path = 'clothing_images'
directory = os.fsencode(dir_path)
```

```python
for file in tqdm(os.listdir(directory)):
    filename = os.fsdecode(file)
    try:
        ses.add_image(os.path.join(dir_path, filename))
    except:
        os.remove(os.path.join(dir_path, filename))
```

```python
deleted = 0

for file in tqdm(os.listdir(directory)):
    filename = os.fsdecode(file)
    relative_path = os.path.join(dir_path, filename)
    try:
        close_images = ses.search_image(relative_path)
        for image in close_images:
            if image['path'] != relative_path:
                if image['dist'] < 0.2:
                    try:
                        deleted += 1
                        os.remove(relative_path)
                    except OSError:
                        deleted -= 1
                        pass
    except:
        pass
    
#     ses.add_image(os.path.join(dir_path, filename))
#     print(filename)
```

```python
# delete not jpg
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    relative_path = os.path.join(dir_path, filename)
    if filename.split('.')[-1] != 'jpg':
        os.remove(relative_path)
```

```python
jpg_downloaded = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.split('.')[-1] == 'jpg':
        jpg_downloaded.append(filename.split('.')[0])
        
```

```python
from os import listdir
from os.path import isfile, join
import pandas as pd
mypath = 'clothing_images'
downloaded = [f.split('.')[0] for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
data_clean = pd.read_csv('clothing_data_asin.csv', index_col=0)
```

```python
data_similar_removed = data_clean.loc[downloaded]
```

```python
data_similar_removed.to_csv('data_similar_removed.csv')
```

```python
jpg_data_similar_removed = data_clean.loc[jpg_downloaded]
```

```python
jpg_data_similar_removed.to_csv('data_similar_removed_jpgs.csv')
```
