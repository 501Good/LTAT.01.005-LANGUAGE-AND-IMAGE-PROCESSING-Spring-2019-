This script is for parsesing and downloading the Amazon Review dataset.

```python
import pandas as pd
import gzip
from itertools import chain
import urllib.request
```

```python
def parse(path):
  g = gzip.open(path, 'r')
  for l in g:
    yield eval(l)
```

```python
data = pd.DataFrame(columns=['asin', 'title', 'price', 'imUrl', 'categories'])
# categories = ['Clothing, Shoes & Jewelry', 'Clothing']
data_generator = parse('data/metadata.json.gz')

i=0

asin_list = []
title_list = []
price_list = []
imUrl_list = []
categories_list = []

for item in data_generator:
    i+=1
    if i%100000 == 0:
        print('%i items'%i)
        print('%i items in list'%len(asin_list))
#     item = next(data_generator)
    cats = list(chain.from_iterable(item.setdefault('categories', [])))
    if any(x == 'Clothing' for x in cats):
        asin_list.append(item['asin'])
        title_list.append(item.setdefault('title', '')),
        price_list.append(item.setdefault('price', 0)),
        imUrl_list.append(item.setdefault('imUrl', '')),
        categories_list.append(';'.join(cats))
#         data.loc[len(data)] = row
```

```python
data = pd.DataFrame({'title': title_list, 
                     'price': price_list, 
                     'imUrl': imUrl_list,
                    'categories': categories_list}, index = asin_list)
```

```python
# download images

i=0
for asin, item in data_clean.iterrows():
    i+=1
    if asin not in downloaded:
        try:
            
            urllib.request.urlretrieve(item.imUrl, "clothing_images/%s.%s"%(asin, item.imUrl.split('.')[-1]))
            downloaded.append(asin)
        except:
            i-=1
            pass
    if i%1000==0:
        print('%i images downloaded'%(i))
```
