```python
vocab = vectorizer.vocabulary_
# inv_vocab = {}
for special_word in ['<start>', '<end>']:
    vocab[special_word] = len(vocab)+1
for key, num in vocab.items():
    vocab[key] = int(num+1)
#     inv_vocab[int(num)] = key
```
